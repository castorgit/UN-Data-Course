#### Transforms World development file trasposing rows with year-indicator series to columns
#### Can be used for any indicator, year 

```python
path = 'E:/$ Deep Learning Courses/$ 2021 06 01 ONU Courses/# WIP/datasets/UN/world-development/'
data = pd.read_csv(path+'indicators.csv')
```

```python
ind_code0 = 'SP.DYN.LE00.MA.IN'
ind_code1 = 'SP.DYN.LE00.FE.IN'
ind_code2 = 'SP.DYN.LE00.IN'
data.columns

#SP.DYN.LE00.FE.IN
#SP.DYN.LE00.MA.IN
#SP.DYN.LE00.IN
df = data
df = df[(df["IndicatorCode"]==ind_code0) | (df["IndicatorCode"]==ind_code1) |  (df["IndicatorCode"]==ind_code2)]
df = df[df['Year']==2010]
df = df.drop('Year', axis=1) 
df = df.drop('IndicatorName', axis=1) 
df1 = df.set_index(['CountryName','CountryCode', 'IndicatorCode','Value'], drop=True).unstack('IndicatorCode').reset_index()
df
```

```python
pivoted = df.pivot(index='CountryName', columns='IndicatorCode', values='Value')\
            .reset_index()
```

```python
pivoted.to_csv(path+'Life_expectancy_at_birth_2010.csv')
```
