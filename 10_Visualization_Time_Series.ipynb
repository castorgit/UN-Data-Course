```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from matplotlib.ticker import FuncFormatter
from matplotlib.dates import DateFormatter, MonthLocator
import matplotlib.dates as mdates
```

```python
file = "./data/Births_in_Canada_2000_2022.xlsx"
df =  pd.read_excel(file, sheet_name='Sheet',  index_col=[0], header=[0])

df.reset_index(inplace=True)

df['Date'] = pd.to_datetime(df['Year of birth'].astype(str) + '-' + df['Month of birth'].astype(str) + '-01')
df
```

```python
# (1) Births per month Canada
# Source Statistics Canada

ts1 = df.copy()

ts1.set_index('Date',inplace=True)

fig, ax = plt.subplots(figsize=(24,8))

plt.title('Births per Year in Canada', fontsize=24)


sns.lineplot(data=ts1, x='Date', y='Number of live births', markers=True, dashes=False, linewidth=5, 
             alpha=0.9, label='Number of live births')



plt.fill_between(ts1.index, ts1['Number of live births'], color="palegreen", alpha=0.3)


ax.grid(which='minor', alpha=0.8)  
ax.set_facecolor('whitesmoke')
plt.grid()
plt.legend(loc="upper right", fontsize=20)
plt.tight_layout()
plt.tick_params(labelsize=20)

#ax.xaxis.set_major_locator(MonthLocator(interval=3))
#ax.xaxis.set_major_formatter(DateFormatter("%b %Y"))
ax.xaxis.set_major_locator(mdates.YearLocator())  # Ticks at January of every year
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))


plt.xticks(rotation=45)

def thousands_formatter(x,pos):
    return f'{x:,.0f}'

# Set the formatter for the y-axis
ax.yaxis.set_major_formatter(FuncFormatter(thousands_formatter))
#plt.savefig('./data/figures/EDA3-04.png')
plt.ylabel("Number of live Births", fontsize=20)
plt.xlabel("Date", fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(24000, 35000)
plt.tight_layout()
plt.savefig('./figures/01_Births_in_Canada.png')
plt.show()
```
